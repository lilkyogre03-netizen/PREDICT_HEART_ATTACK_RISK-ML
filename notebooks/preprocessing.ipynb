```python
!pip install pandas
import pandas as pd


df=pd.read_csv("D:\DATA_Science\HEART ATTACK ML\data\Raw\heart_attack_dataset.csv")

df["cholesterol"]=df["cholesterol"].fillna(round(df["cholesterol"].median()))
df["smoking_status"]=df["smoking_status"].fillna("unknown")
df["physical_activity"]=df["physical_activity"].fillna("unknown")
df["alcohol_consumption"]=df["alcohol_consumption"].fillna("unknown")

df["resting_bp"]=df["resting_bp"].fillna(df["resting_bp"].median())
df["max_heart_rate"]=df["max_heart_rate"].fillna(df["max_heart_rate"].median())
df["oldpeak"]=df["oldpeak"].fillna(df["oldpeak"].median())
df["bmi"]=df["bmi"].fillna(df["bmi"].median())
df["stress_level"]=df["stress_level"].fillna(df["stress_level"].median())
df["fasting_blood_sugar"] = df["fasting_blood_sugar"].fillna(
    df["fasting_blood_sugar"].mode()[0]
)

df["alcohol_consumption"]=df["alcohol_consumption"].map({
      "Non-drinker": 0, "Moderate": 1, "Heavy": 2, "Other": 3,"unknown": 3
})
df["physical_activity"] = df["physical_activity"].map({
    "Low": 0, "Moderate": 1, "High": 2, "Other": 3, "unknown": 3
})
df["thalassemia"] = df["thalassemia"].map({
    "Normal": 0,
    "Reversible Defect": 1,
    "Fixed Defect": 2
})
df["gender"] = df["gender"].map({
    "Male":0 , "Female":1
})
df = pd.get_dummies(df, columns=["resting_ecg", "chest_pain_type"], dtype=int)
df = df[df["smoking_status"] != "unknown"]
df = pd.get_dummies(df, columns=["smoking_status", "st_slope"], dtype=int)

df=df.drop(columns="patient_id")
df.to_csv("D:/DATA_Science/HEART ATTACK ML/data/processed/heart_clean.csv", index=False)

# corr_matrix = df.corr(numeric_only=True)
# plt.figure(figsize=(10, 8)) 
# sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
# df= df.drop(columns = "stress_level") 
# plt.title('Correlation Heatmap')
# plt.show()
```
